```python
## which of these correlations are truly significant, though?
import scipy.stats

import numpy as np

NUM_COUNTIES = 2600

NUM_STATES = 50

r = 0.4 ## is .1 stat sig?

denom = (1 - (r**2))

t = r * np.sqrt((NUM_COUNTIES - 2)/ denom)

county_crit = scipy.stats.t.ppf(q=1 - .01/2, df=NUM_COUNTIES)

print(f't for counties is {t}, {county_crit}')

t = r * np.sqrt((NUM_STATES - 2)/ denom)

state_crit = scipy.stats.t.ppf(q=1 - .01/2, df=NUM_STATES)


print(f't for states is {t}, critical is {state_crit}')



```
